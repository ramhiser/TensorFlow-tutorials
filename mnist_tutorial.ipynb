# Softmax Regression Model
As in the original tutorial, we begin with a simple model:

$$y = \text{softmax}(\textbf{Wx} + \textbf{b}),$$

where

$$
    \text{evidence}_i = \sum_j W_{i,~ j} x_j + b_i\\
    \text{softmax}(x_i) = \frac{\exp(x_i)}{\sum_j \exp(x_j)}
$$
